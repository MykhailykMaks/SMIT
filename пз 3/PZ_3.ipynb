## Мета: набути практичних навичок у розв’язанні задач з підрахунку ймовірностей на основі геометричного визначення ймовірності, алгебри подій та теорем множення і додавання ймовірностей; навчитись застосовувати на практиці формули повної ймовірності та Байєса.

## 1. Парадокс Монті Хола. «Уявіть себе на телегрі, де вам потрібно обрати одну з трьох дверей: за одними з них автомобіль; за двома іншими по козі. Ви обираєте одні двері, наприклад, перші, ведучий відчиняє одні з двох інших, наприклад, треті, за якими коза. Тоді він каже вам: «Бажаєте змінити вибір на другі двері?» Чи отримаєте ви перевагу, якщо зміните свій вибір?» 

### Коли ми вибираємо перші двері, ймовірність виграти 1/3, ведучий відкриває двері в яких немає автомобіля в любому випадку, але йомвірність нашого програшу всеодно не змінилася, отже, якщо ми не змінимо наш вибір ймовірність на перемогу залишиться 1/3, а якщо змінимо то 2/3.

## 2. У 2 урнах лежать кульки, що відрізняються лише кольором. В першій урні 5 білих, 11 чорних, та 8 красних кульок, в другій – відповідно 10, 8 та 6. З oбох урн наугад витягнуто по одній кульці. Знайти ймовірність того, що витягнуті кульки одного кольору

## 3. В кожній з 3 урн по 6 чорних та 4 білих кульки. З першої урни наугад витягнута одна кулька та перекладена у другу урну. Після цього з другої урни наугад витягнута одна кулька та перекладена у третю урну. Знайти ймовірність того, що кулька, що витягнута потім з третьої урни, буде білою.

## 4. 4 стрілка незалежно один від одного стріляють по одній мішені,  роблячи по одному пострілу. Ймовірності влучення для кожного стрілка складає 0.4, 0.6, 0.7, 0.8. Після стрільби встановлено, що у мішень влучили 3 рази. Знайти ймовірність того, що не влучив четвертий стрілок.

## 5. Батарея з трьох гармат зробила залп, причому два снаряди влучили в мішень. Знайти ймовірність того, що перша гармата дала влучення, якщо ймовірності влучення у мішень першою, другою та третьою гарматою складають відповідно 0.4, 0.3, 0.5.
