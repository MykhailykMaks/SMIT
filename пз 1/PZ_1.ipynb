# Практична робота №1

## Тема: Елементи комбінаторики. Класичне визначення ймовірності. Застосування комбінаторики для розрахунку ймовірностей

## Мета: Набути практичних навичок у розв’язанні задач з комбінаторики.

## 1. Знайдіть число можливих «слів» з літер слова «зоологія». Скільки таких слів, в яких 3 літри «о» розташовані поряд?

## 2. Маємо 20 найменувань товару. Скількома способами їх можна розподілити по 3 магазинах, якщо відомо, що в перший магазин має бути доставлено 8 найменувань, у другий – 7 найменувань і в третій – 5 найменувань товару? 

## 3. Емма хоче купити сонячні окуляри. В магазині окулярів є такий вибір: купити готові сонячні окуляри, або замовити, скомбінувавши оправу з лінзами. Для створення власної моделі Емма має вибір з двох оправ та двох брендів лінз, кожен з яких пропонує чотири виді лінз. Серед готових окулярів є п’ять доступних моделей. Скільки всього варіантів придбати окулярі є у Емми?

## 4. Скільки словників потрібно видати, щоб можливо було безпосередньо виконати переклади з будь-якої з п’яти мов: російської, англійської, французької, німецької, італійської – на будь-яку з цих п’яти мов?

## 5. Скількома способами на шаховій дошці можливо вказати:
## а) 2 клітинки?
## б) 2 клітинки одного кольору?
## в) 2 клітинки різного кольору?

## Контрольні запитання

### 1. Комбінаторика зосереджена на підрахунку можливих комбінацій, перестановок і розміщень, які можна утворити з цих елементів, керуючись певними правилами.

### 2. Це модель для ілюстрації комбінаторних задач, за допомогою якої можна легко візуалізувати складні концепції, такі як вибірка з поверненням або без повернення. Вона має велике значення для комбінаторики, так як є легкою для уявлення та візуалізаціїї, а за її допомогою можна легко вирішувати задачі.

### 3. Перестановка - це спосіб упорядкування елементів із заданої множини.

### 4. n!/(n-k)! або n^k, в залежності від того чи є в умові повторення елементів.

### 5. n!/(k!*(n-k)!)
