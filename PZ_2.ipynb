# Практична робота №2

## Тема: Класичне визначення ймовірності. Застосування комбінаторики для розрахунку ймовірностей.

## Мета: Набути практичних навичок у розв’язанні задач з підрахунку ймовірностей на основі класичного визначення з використанням формул комбінаторики.

## 1. В урні 6 білих та 4 чорних кульки. З цієї урни навмання взято 5 кульок. Знайти ймовірність того, що 2 з них білі, а 3 – чорні.

## 2. В урні 10 кульок, з яких 2 білих, 3 чорних та 5 синіх. Навмання взято 3 кульки. Знайти ймовірність того, що всі 3 кульки різного кольору.

## 3. На 10 картках написані літери «а», «а», «а», «м», «м», «т», «т» , «е», «и», «к». Картки ретельно перемішано та викладено у ряд.
## Знайти ймовірність того, що отримаємо слово «математика».

## 4. На п’ятимісну лавку випадковим чином сідають 5 людей. Знайти ймовірність того, що певні 3 людини будуть сидіти поруч

## 5. В урні 10 кульок. Ймовірність того, що 2 взяті кульки будуть білими, складає 2/15. Скільки в урні білих кульок?

## Контрольні запитання

### 1. Це відношення кількості сприятливих наслідків, які сприяють настанню події A, до загальної кількості всіх рівноможливих наслідків певного випробування.

### 2. Це будь-яка дія, процес або спостереження, результат якого неможливо точно передбачити заздалегідь.

### 3. Комбінаторика використовується як «механізм підрахунку» у класичній теорії ймовірностей.

### 4. Принципова відмінність полягає в умовах застосування та математичній строгості визначень. Класичне визначення є спрощеним, тоді як ймовірність на просторі елементарних подій є частиною більш загальної та математично строгої аксіоматичної теорії ймовірностей.

### 5. Беремо по одній букві: Взяти потрібну першу букву ймовірність - 1/5, другу - 1/4 і так до 1/1, оскільки нам потрібні всі 5 букв, отже всі ймовірності перемножуємо:(1/5*1/4*1/3*1/2*1).Результат 1/120
